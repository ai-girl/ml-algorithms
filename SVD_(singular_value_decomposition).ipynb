**Singular Value Decomposition (SVD)** can be looked at from three mutually compatible points of view.

On the one hand, we can see  it as a method for transforming  correlated variables into a set of uncorrelated ones that better expose the various relationships among the original data items. 

At the same time, <font color='darkblue'>SVD is a method for identifying and ordering the dimensions along which data points exhibit the most variation.</font>

This ties in to the third way of viewing  **SVD**, which is that once we have identified where the most variation is, it's possible to find the best approximation of the original data points using fewer dimensions.

Hence, **SVD** can be seen as a method for data reduction.

The regression line running through them shows the best approximation of the original data with a 1-dimensional object (a line). It is the best approximation in the sense that it is the line that minimizes the distance between each original point and the line. 

If we drew a perpendicular line from each point to the regression line, and took the intersection of those lines as the approximation of the original datapoint, we would have a reduced representation of the original data that captures as much of the original variation as possible. 

Notice that there is a second regression line, perpendicular to the first, shown in Figure 2. This line captures as much of the variation as possible along the second dimension of the original data set. 

It does a poorer job of approximating the orginal data because it corresponds to a dimension exhibiting less variation to begin with. It is possible to use these regression lines to generate a set of uncorrelated data points that will show subgroupings in the original data not necessarily visible at first glance.

### Basic Idea of SVD 

<font color='darkblue'>Taking a high dimensional, highly variable set of data points and reducing it  to a lower dimenisonal space that exposes the substructure of the original data more clearly and orders it from the most variation to the least.</font>

## Example of a Full Singular Value Decomposition

**SVD** is based on a theorem from linear algebra which says that a rectangular matrix $A$ can be broken down into the product of three matrices <font color='darkblue'>
- An Orthogonal Matrix $U$
- A diagonal Matrix $S$
- Transpose of Orthogonal Matrix $V$.</font>

The theorem is presented is usually, like :      

\begin{align}
A_{mn} = U_{mm}S_{mn}V_{n,n}^T
\end{align}

where $U^TU = I$, $V^TV=I$.

- The columns of **$U$** are **orthonormal eigenvectors of $AA^T$**
- The columns of **$V$** are **orthonormal eigenvectors of $A^TA$**
- **$S$**  is a **diagonal matrix containing the square roots of Eigenvalues** of $U$ or $V$  in descending order.  

*Steps in Solving SVD in form of an algorithm :*

```
Algorithm :    
 -------------------------------------------------------------------
  Input : A rectangular matrix A (m*n)
  Output : 
  1. Orthonormal Matrix U(m*m)
  2. Diagonal Matrix S(m*n)
  3. Transposed Orthogonal Matrix V (n*n)
-------------------------------------------------------------------
Step 1. Calculate AA^T
        (Since AA^T is symmetric and Positive Semi-Definite Matrix)
Step 2. Calculate the Eigen values and Eigen Vectors of AA^T
Step 3. Arrange Eigen vectors into column vectors of U, 
        in order of the magnitude of the Eigen Value. 
        (Eigen Vectors corresponding to highest maginitude Eigen value, 
        comes first, then next and so no)
Step 4. Convert the Eigen vectors into Orthonormal vectors, using 
        Gram Schmidt Orthogonalization.
Step 5. Calculate A^TA
Step 6. Repeat Steps 2,3,4 for A^T A
Step 7. Diagonal Matrix S, consists of square root of Eigen values, in descending order of magnitude. 
        Return U,S, and V^T
```

<font color='Blue'>

$_1u_1v_1$ is the largest.

</font>
What we want from the SVD ?
The right bases for the four subspaces. 
Then we describe the steps to find the bases vectors  **in order of importance**

- We have to have two sets of singular vectors $u's$ and $v's$. The $u's$ are in $^m$ and the $v's$ are in $^n$. which will be columns of $m \times m$ matrix $U$ and a $n$ by $n$ matrix $V$.

(using Vectors) The $u's$ and $v's$ give bases for the four fundamental subspaces :      
<font color='Blue'>

1. $v_1,\ldots,v_r$ is an orthonormal basis for **row space**
1. $v_{r+1},\ldots,u_n$ is an orthonormal basis for **nullspace**

## Solve SVD - Using Power Iteration Algorithm 

Source : [Power Iteration](https://www.wikiwand.com/en/Power_iteration)

In mathematics, **power Iteration** ( also known as Power Method) is an Eigen Value algorithm: given a diagonalizable matrix $A$, the algorithm will produce $$, which is the greatest Eigen value of $A$, and a nonzero eigenvector $v$, which is corresponding to the eigenvalue $$., that is $Av = v$.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
def eigen_value(B,b_k):

  # uses property A.v = lamda .v 
  lambda_ = np.linalg.norm(np.dot(B,b_k))
  return lambda_
````

````python
def power_iteration(A,eps):

  B = np.dot(A.T, A)
  b_k = np.random.randn(A.shape[1])
  current = b_k

  while True:
    
    b_k1 = np.dot(B,b_k)
    b_k1_norm  = np.linalg.norm(b_k1)
    b_k = np.divide(b_k1,b_k1_norm)
    next = b_k
    
    if abs(np.dot(current, next)) > 1 -  eps:
      S = np.sqrt(eigen_value(B,b_k))
      return S,b_k   
    current  = b_k
````

````python
matrix_A = np.array([[3,1,1],[-1,3,1]])
eps = 1e-10
eigen_value_result, eigen_vector = power_iteration(matrix_A,eps)

print(eigen_value_result)
print(eigen_vector)
````

````python
u,s,v = np.linalg.svd(matrix_A)
print(s)
print(v)
````

# Convergence Of The  Power Method 

Theorem : If $A$ is diagonalizable and has dominant eigenvalue. Then Power Iteration sequence $Ax, A^2x, A^3x,\ldots, A^kx$ converges to the dominant eigen value. 

Proof: Since $A$ is diagonalizable, it has $n$ linearly independent eigenvectors $v_1,\ldots, v_n$ and corresponding eigenvalues $_1,\ldots, _n$

- Let $v_1$ and $_1$ be dominant
- because the eigen vectors are independent they form a basis. 
    - so any vector $x_0$ can be represented as :       
      $x_0 = c_1 v_1 \ldots+ c_n v_n$
    - And $c_1$ must be non-zero, otherwise $x_0$ is not a good choice, we need to choose another initial vector.
- let's multiply both sides by $A$
- $Ax_0 = c_1 A v_1 \ldots+ c_n A v_n$\
 $ \text{    } \text{   }\text{ } \text{  }\text{ }\text{ }\text{ }   = c_1 _1 v_1 \ldots+ c_n _n v_n$

Repeat it $k$ times\
$  \text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    } \text{   }\text{ } \text{  }\text{ }\text{ }\text{ } = c_1 _1^k v_1 \ldots+ c_n _n^k v_n$\
$  \text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    }\text{    } \text{   }\text{ } \text{  }\text{ }\text{ }\text{ } = c_1 _1^k( c_1 v_1 +c_2 (\frac{_2}{_1})^k\ldots+ c_n (\frac{_n}{_1})^k v_n$

Since $_1$ is dominating, the ratios $(\frac{_i}{_1})^k  0$ as $k  $ for all $i$.

So, $A^kx_0 = _1^k c_1v_1$ and it gets better as $k $.

### SVD From Scratch : Greedy Optimization Routine 


````python
def svd(A,k=None,eps=1e-10):

  n,m = A.shape

  svdsofar = []
  S = []
  u_s = []
  v_s = []
  if k == None:
    k = m
  for i in range(k):

    matrix1D = A.copy()

 
    for singularvalue, u, v in svdsofar[:i]:   
      matrix1D = matrix1D - (singularvalue *  np.outer(u,v))

    singularvalue, v =  power_iteration(matrix1D,eps)
    u  = np.dot(A,v)/np.linalg.norm(np.dot(A,v))
    svdsofar.append((singularvalue,u,v))
    # Maintaining seperate lists for U,sigma,V
    S.append(singularvalue)
    u_s.append(u)
    v_s.append(v)

  v_s = np.array(v_s).reshape(k,A.shape[1])
  u_s = np.array(u_s).reshape(k,A.shape[0])
  S = np.array(S).reshape(1,k)
  #SingularVal, Us, Vs = [np.array(x) for x in zip(*svdsofar)]

  return S, u_s, v_s
````

````python
# example 2*3 matrix
A = np.array([[3,1,1],[-1,3,1]])

S,Us,Vs = svd(A,eps=1e-10)

print(S)
print(Us)
print(Vs)
````
